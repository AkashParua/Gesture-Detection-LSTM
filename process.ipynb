```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
#creaating a label manp
label_map = {'back':0, 'forward':1, 'left' :2, 'right':3 , 'wave' :4}
```

```python
import os
import numpy as np
```

```python
squences , labels = [], []
```

```python
import glob
paths = glob.glob('Data/*/*.npy')
for path in paths:
    label = path.split('/')[-2]
    print(label)
    print(path)
    label = label_map[label]
    data = np.load(path)
    squences.append(data)
    labels.append(label)
```

```python
x , labels = np.asarray(squences) , np.asarray(labels)
```

```python
x.shape 
```

```python
y = to_categorical(labels)
```

```python
y.shape
```

Sucessfully Created data

```python
#Using stratified shuuffle split
from sklearn.model_selection import StratifiedShuffleSplit

# Create a StratifiedShuffleSplit object with n_splits=1 and test_size=0.2
split = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=42)
x_test , x_train , y_train , y_test = None , None , None , None
# Use the split() method to generate the train and test indices
for train_index, test_index in split.split(x, y):
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
np.save('x_train.npy' , x_train)
np.save('x_test.npy' , x_test)
np.save('y_train.npy' , y_train)
np.save('y_test.npy' , y_test)
```
